## Juego del Ahorcado
Desarrolla el juego del ahorcado en python. En cada turno irá mostrando por pantalla ```_ _ _ _ _``` las rayas bajas correspondientes al número de letras que tiene la palabra a adivinar. El usuario deberá insertar una letra y, si esa letra existe, se mostrará o mostrarán las letras acertadas ```S _ S _ _``` en sus posiciones correspondientes. 

Si falla, se mostrará el número de errores que lleva y, si lleva x errores, entonces se mostrará "Fin del juego".

1. Inicializa el juego: declara una palabra a adivinar y muestra por pantalla la palabra oculta.


2. Funcionamiento turnos: mediante un bucle genera el sistema de turnos donde el usuario introduce una letra y le va mostrando su progreso y sus vidas.

3. Desarrolla la finalización de la partida, tanto si gana como si pierde.

4. Encapsula el juego en una función llamada ```jugar_ahorcado()``` con la palabra a adivinar y el número de vidas para seleccionar la dificultad como argumentos de entrada.

````python
import random

def ahorcado():
    # Inicialización del juego
    palabra_a_adivinar = "python"  # Palabra secreta a adivinar
    vidas = 6  # Número de intentos o vidas
    letras_adivinadas = ["_"] * len(palabra_a_adivinar)  # Lista para almacenar el progreso de las letras adivinadas
    letras_erradas = []  # Lista para almacenar las letras que el jugador ha fallado
    
    print("Bienvenido al juego del Ahorcado!")
    print("La palabra a adivinar tiene", len(palabra_a_adivinar), "letras.")
    print(" ".join(letras_adivinadas))  # Muestra la palabra oculta como guiones bajos

    # Bucle principal del juego
    while vidas > 0 and "_" in letras_adivinadas:
        print("\nTienes", vidas, "vidas restantes.")
        print("Letras falladas:", ", ".join(letras_erradas))
        letra = input("Introduce una letra: ").lower()
        
        # Verificar si la letra introducida es válida
        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, introduce solo una letra.")
            continue
        
        # Verificar si la letra ya fue adivinada
        if letra in letras_adivinadas or letra in letras_erradas:
            print("Ya has adivinado o fallado esta letra.")
            continue
        
        # Comprobar si la letra está en la palabra
        if letra in palabra_a_adivinar:
            print("¡Acertaste!")
            # Actualizar la lista de letras adivinadas
            for i in range(len(palabra_a_adivinar)):
                if palabra_a_adivinar[i] == letra:
                    letras_adivinadas[i] = letra
        else:
            print("Fallaste.")
            vidas -= 1
            letras_erradas.append(letra)
        
        # Mostrar el progreso
        print(" ".join(letras_adivinadas))
    
    # Condición de victoria o derrota
    if "_" not in letras_adivinadas:
        print("¡Felicidades! Has adivinado la palabra:", palabra_a_adivinar)
    else:
        print("Fin del juego. La palabra era:", palabra_a_adivinar)

# Ejecutar el juego
ahorcado()
################################################################################################

def jugar_ahorcado(palabra_a_adivinar, vidas):
    # Inicialización del juego
    letras_adivinadas = ["_"] * len(palabra_a_adivinar)  # Lista para almacenar el progreso de las letras adivinadas
    letras_erradas = []  # Lista para almacenar las letras que el jugador ha fallado
    
    print("Bienvenido al juego del Ahorcado!")
    print("La palabra a adivinar tiene", len(palabra_a_adivinar), "letras.")
    print(" ".join(letras_adivinadas))  # Muestra la palabra oculta como guiones bajos

    # Bucle principal del juego
    while vidas > 0 and "_" in letras_adivinadas:
        print("\nTienes", vidas, "vidas restantes.")
        print("Letras falladas:", ", ".join(letras_erradas))
        letra = input("Introduce una letra: ").lower()
        
        # Verificar si la letra introducida es válida
        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, introduce solo una letra.")
            continue
        
        # Verificar si la letra ya fue adivinada
        if letra in letras_adivinadas or letra in letras_erradas:
            print("Ya has adivinado o fallado esta letra.")
            continue
        
        # Comprobar si la letra está en la palabra
        if letra in palabra_a_adivinar:
            print("¡Acertaste!")
            # Actualizar la lista de letras adivinadas
            for i in range(len(palabra_a_adivinar)):
                if palabra_a_adivinar[i] == letra:
                    letras_adivinadas[i] = letra
        else:
            print("Fallaste.")
            vidas -= 1
            letras_erradas.append(letra)
        
        # Mostrar el progreso
        print(" ".join(letras_adivinadas))
    
    # Condición de victoria o derrota
    if "_" not in letras_adivinadas:
        print("¡Felicidades! Has adivinado la palabra:", palabra_a_adivinar)
    else:
        print("Fin del juego. La palabra era:", palabra_a_adivinar)

# Ejemplo de ejecución
jugar_ahorcado("python", 6)  # Llamada al juego con la palabra 'python' y 6 vidas

````
